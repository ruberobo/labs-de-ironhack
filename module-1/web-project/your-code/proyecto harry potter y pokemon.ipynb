```python
# alguien ya instaló los api de pokemon a python
import pokepy

#seleccionó un pokemon al azar

x = random.choice(pokedata['name'])
x
```

pokepy documentation: https://pokeapi.github.io/pokepy

```python
def describe_pokemon(x):
    pokemon = pokepy.V2Client().get_pokemon(x)
    
  
    #status
    a = []
    b = []

    for i in range(len(pokemon.stats)):
        a.append(pokemon.stats[i].stat.name)
        b.append(pokemon.stats[i].base_stat)
        status = dict(zip(a,b))
    
    #tipos
    types = []
    for i in range(len(pokemon.types)):
        types.append(pokemon.types[i].type.name)
    
  
    
    #habilidades
    abilities = []
    for i in range(len(pokemon.abilities)):
        abilities.append(pokemon.abilities[i].ability.name)
    
    
    return f' name {pokemon.name}, status {status}, types {types}, abilities {abilities}'
```

```python
describe_pokemon(x)
```

```python
#la funcion tambien funciona con el número o el nombre del pokemon como string
describe_pokemon(666)
```
