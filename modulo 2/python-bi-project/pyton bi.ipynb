Selecting a place World Wide for health company

```python
#se elig salud porque es el tema d inters
salud = list(ofi_colec.find({'$or':[{'category_code':'health'},
                             {'category_code':'medical'},
                             {'category_code':'hospitality'}]},
                             {'_id':0,'name':1,'offices.country_code':1, 'offices.state_code':1 ,
                                                        'offices.latitude':1,'offices.longitude':1,
                                                        'number_of_employees':1,'competitions':1})
         .sort('number_of_employees',-1))
```

```python
salud
```

```python
#ordeno la salud

salud1=[]
for i in range(len(salud)):
    for j in salud[i]['offices']:
        j['competencia'] = salud[i]['competitions']
        j['name']=salud[i]['name']
        j['empleados'] = salud[i]['number_of_employees']
        salud1.append(j)
salud1
```

```python
salud1df=pd.DataFrame(salud1)
salud1df
#the top three of number of employees are on USA
```

```python
import matplotlib.pyplot as plt
salud1df['state_code'].value_counts()
#california has the most companies and NY has the top of number of employess
```

```python
#Mapa
mapa2=folium.Map(location=[37, -95], zoom_start=5) #coordenadas de usa
```

```python
#agrego circulos
for i in range(len(salud1df['latitude'])):
    try:
        folium.Circle([salud1df['latitude'][i],salud1df['longitude'][i]], 1000, fill = True).add_child(folium.Popup(salud1df['name'][i])).add_to(mapa2)
    except:
        pass
    
mapa2
```
