# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

```python
data.info()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data1= data.groupby('ItemType')['RetailSales','RetailTransfers','WarehouseSales'].agg(['sum'])
data1
```

```python

data1.plot(kind = 'bar')
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data['total_sales']=data['RetailSales']+data['WarehouseSales'] #sin transfers
```

```python
data2 = data.pivot_table(data, index=['Supplier'])
```

```python
data21= data2.sort_values('total_sales',ascending=False ).head(10)
data21
```

```python
data21['total_sales'].plot(kind = 'barh', title='top 10 suppliers')
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
data3 = data.pivot_table(data, index=['Year','Month'])
data3
```

```python
data3[['RetailSales','RetailTransfers','WarehouseSales']].plot(kind = 'line', title='sales throuh time')
```

## 4. Plot the same information as above but as a bar chart.

```python
data3[['RetailSales','RetailTransfers','WarehouseSales']].plot(kind = 'bar', title='sales throuh time')
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
data4= data.groupby(['Year','Month','ItemType'])['RetailSales'].agg(['sum'])
data4
```

```python
data5 = data.pivot_table(data, index=['Year','Month'], columns = 'ItemType')
data5
```

```python
data5['RetailSales'].plot(kind = 'line', title='Retail sales throuh time')
```

## 6. Plot the same information as above but as a bar chart.

```python
data5['RetailSales'].plot(kind = 'bar', title='Retail sales throuh time')
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
x = data5['RetailSales'][['BEER','DUNNAGE','KEGS','LIQUOR','NON-ALCOHOL','REF','STR_SUPPLIES','WINE']]
y = data5['RetailTransfers'][['BEER','DUNNAGE','KEGS','LIQUOR','NON-ALCOHOL','REF','STR_SUPPLIES','WINE']]
```

```python
for i,j in zip(x,y):
    plt.scatter(x[i],y[j], label = i)
    plt.xlabel("Retail Sales")
    plt.ylabel("Retail Transfers")

plt.legend(loc='upper right')
plt.show();
```

```python
#metodo de https://stackoverflow.com/questions/4270301/matplotlib-multiple-datasets-on-the-same-scatter-plot
x = data5['RetailSales'][['BEER','DUNNAGE','KEGS','LIQUOR','NON-ALCOHOL','REF','STR_SUPPLIES','WINE']]
y = data5['RetailTransfers'][['BEER','DUNNAGE','KEGS','LIQUOR','NON-ALCOHOL','REF','STR_SUPPLIES','WINE']]
fig = plt.figure()
ax1 = fig.add_subplot()

for i,j in zip(x,y):
    ax1.scatter(x[i], y[i], label = i)

    
    
plt.xlabel("Retail Sales")
plt.ylabel("Retail Transfers")    
plt.legend(loc='upper right')
plt.show();
```
