```python
from pymongo import MongoClient
client=MongoClient()
db=client.companies
```

```python
db.collection_names()
```

```python
colec = db.companies
```

```python
list(colec.find().limit(10))
```

# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
list(colec.find({'name':'Babelgum'},{'_id':0,'name':1,'email_address':1}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
list(colec.find({'number_of_employees':{'$gt':5000}},
                {'name':1,'_id':0,'number_of_employees':1}).sort('number_of_employees',-1).limit(20))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
m= colec.find({'$and':[{'founded_year':{'$gte':2000}},
                       {'founded_year':{'$lte':2005}}]},
              {'name':1,'founded_year':1,'_id':0}).sort('founded_year',-1)
```

```python
list(m)
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
list(colec.find({'$and':[{'ipo.valuation_amount':{'$gt':100000000}},
                       {'founded_year':{'$lte':2010}}]},
              {'name':1,'founded_year':1,'_id':0}).sort('founded_year',-1))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
p5 = colec.find({'$and':[{'number_of_employees':{'$lt':1000}},
                   {'founded_year':{'$lt':2005}}]},
           {'name':1, 'founded_year':1}).sort('number_of_employees').limit(10)
```

```python
list(p5)
```

### 6. All the companies that don't include the `partners` field.

```python
p6 = (colec.find({'partners':{'$exists':False}})).limit(10)
```

```python
list(p6) #no salio nada 
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
p7 = (colec.find({'category_code':None})).limit(10)
```

```python
list(p7)
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
p8 = colec.find({'$and':[{'number_of_employees':{'$gt':100}},
                   {'number_of_employees':{'$lt':1000}}]},
           {'name':1, 'number_of_employees':1}).sort('number_of_employees').limit(10)
```

```python
list(p8)
```

### 9. Order all the companies by their IPO price descendently.

```python
 p9 = (colec.find({},{'ipo.valuation_amount':1,'_id':0,'name':1,}).sort('ipo.valuation_amount',-1)).limit(20)
```

```python
list(p9)
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
 p10 = (colec.find({},{'number_of_employees':1,'_id':0,'name':1,}).sort('number_of_employees',-1)).limit(10)
```

```python
list(p10)
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
 p11 = colec.find({'founded_month':{'$gte':7}},
                       {'_id':0,'name':1, 'founded_month':1}).limit(1000)
```

```python
list(p11)
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
p12 = colec.find({'deadpooled_year':{'$gte':3}},
                       {'_id':0,'name':1, 'deadpooled_year':1}).limit(20)
```

```python
list(p12)
```

### 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
p13 = colec.find({'$and':[{'founded_year':{'$lt':2000}},
                   {'acquisition.price_amount':{'$gt':10000}}]},
           {'name':1, 'acquisition.price_amount':1,'founded_year':1}).limit(20)
```

```python
list(p13)
```

### 14. All the companies that have been acquired after 2015, order by the acquisition amount, and retrieve only their `name` and `acquisiton` field.

```python
 p14 = (colec.find({'acquired_year':{'$gt':2015}},{'name':1, 'acquisition':1}).sort('acquisition.price_amount',-1)).limit(10).limit(20)
```

```python
list(p14)
```

### 15. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
 p15 = (colec.find({},{'name':1,'_id':0,'founded_year':1,}).sort('founded_year',-1)).limit(10)
```

```python
list(p15)
```

### 16. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `aquisition price` descendently. Limit the search to 10 documents.

```python
p16 = colec.find({"founded_day":{'$lte': 7}},
                {'_id':0,'name':1,"founded_day":1}).sort('acquisition.price_amount',1).limit(10)
```

```python
list(p16)
```

### 17. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascendant order.

```python
p17 = colec.find({"$and":[{"category_code":"web"},{'number_of_employees':{'$gt':4000}},]},
                {'_id':0,'name':1,'number_of_employees':1}).sort('number_of_employees',1)
```

```python
list(p17)
```

### 18. All the companies which their acquisition amount is more than 10.000.000, and currency are 'EUR'.


```python
p18 = colec.find({"acquisition.price_currency_code":"EUR",
                           'acquisition.price_amount':{'$gt':10000000}},
                {'_id':0,'name':1,'acquisition.price_amount':1}).sort('acquisition.price_currency_code',1).limit(20)
```

```python
list(p18)
```

### 19. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
p19 = colec.find({'acquisition.acquired_month':{'$lt':3}}, {'_id':0, 'name':1, 'acquisition':1}).limit(10)
```

```python
list(p19)
```

### 20. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
p20 = colec.find({"founded_year":{"$gte":2000,"$lte":2010},
                  'acquisition.acquired_year':{"$lte":2011}},
                  {'_id':0, 'name':1, 'founded_year':1, 'acquisition.acquired_year':1}).limit(10)
```

```python
list(p20)
```
