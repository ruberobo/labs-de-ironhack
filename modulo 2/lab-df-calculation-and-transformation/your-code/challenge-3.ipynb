# Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, import `Orders.csv` from the "subsetting" lab folder into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# enter your code here
orders = pd.read_csv("../../lab-subsetting-and-descriptive-stats/your-code/Orders.csv")
orders.head()
```

```python
orders.info()
```

```python
VIP = orders[orders['amount_spent']>orders['amount_spent'].quantile(0.95)]
VIP
```

```python
'''UK is the country with more VIP customers'''
VIP['Country'].value_counts()
```

```python
Preferred = orders[(orders['amount_spent']==((orders['amount_spent'].quantile(0.95)) -
                    (orders['amount_spent'].quantile(0.75))))]
               
Preferred.info()
```

```python
'''UK is the country with more preferred customers'''
Preferred['Country'].value_counts()
```
