## Regression Model Evaluation

```python
from sklearn.datasets import load_boston

data = load_boston()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.DataFrame(data["target"], columns=['MEDV'])

boston = pd.concat([X, y], axis=1)
```

```python
boston
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test, y_train, y_test = tts(boston.drop('MEDV', axis=1),boston['MEDV'])
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

```python
from sklearn.linear_model import LinearRegression as LinReg
```

```python
linreg = LinReg()# Se asinga el modelo a una variable 
```

```python
linreg.fit(x_train,y_train) # Se entrena el modelo con x_Train y y_train
```

```python
y_pred=linreg.predict(x_test) #Se hace la prediccion con x_test
y_pred2=linreg.predict(x_train)
```

## 3. Calculate and print R-squared for both the training and the testing set.

```python
from sklearn.metrics import r2_score as r2

r2(y_test, y_pred)
```

```python
r2(y_train, y_pred2)
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```

```python
mean_squared_error(y_train, y_pred2)
```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_error(y_train, y_pred2)
```

## Classification Model Evaluation

```python
from sklearn.datasets import load_iris

data = load_iris()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.DataFrame(data["target"], columns=["class"])

iris = pd.concat([X, y], axis=1)
```

```python
iris
```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
x_train,x_test, y_train, y_test = tts(iris.drop('class', axis=1),iris['class'])
```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
from sklearn.linear_model import LogisticRegression
logit_reg = LogisticRegression()
logit_fit = logit_reg.fit(x_train,y_train)
y_pred = logit_fit.predict(x_test)
y_pred2 = logit_fit.predict(x_train)
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
from sklearn.metrics import accuracy_score as acc

acc(y_test, y_pred) 
```

```python
acc(y_train, y_pred2)
```

## 9. Calculate and print the balanced accuracy score for both the training and the testing set.

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

## 10. Calculate and print the precision score for both the training and the testing set.

```python
from sklearn.metrics import precision_score as prec

prec(y_test, y_pred, average='macro')
```

## 11. Calculate and print the recall score for both the training and the testing set.

```python
from sklearn.metrics import recall_score as rec

rec(y_test, y_pred, average='macro')
```

```python
rec(y_train, y_pred2, average='macro')
```

## 12. Calculate and print the F1 score for both the training and the testing set.

```python
from sklearn.metrics import f1_score as f1

f1(y_test, y_pred, average='macro')
```

```python
f1(y_train, y_pred2, average='macro')
```

## 13. Generate confusion matrices for both the training and the testing set.

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
confusion_matrix(y_train, y_pred2)
```

## Bonus: For each of the data sets in this lab, try training with some of the other models you have learned about, recalculate the evaluation metrics, and compare to determine which models perform best on each data set.

```python
#funcion clase de hoy
import re
def modeling_testing(lista_modelos, data, target):
    X_train, X_test, y_train, y_test = tts(data.loc[:, data.columns != target], data[target])
    #importo
    from sklearn.metrics import accuracy_score as acc
    from sklearn.metrics import balanced_accuracy_score
    from sklearn.metrics import precision_score as prec
    from sklearn.metrics import recall_score as rec
    from sklearn.metrics import f1_score as f1
    from sklearn.metrics import confusion_matrix
    
    for i in lista_modelos:
        modelo = i()
        modelo.fit(X_train, y_train)
        y_pred = modelo.predict(X_test)
        #pruebas de la clase 
        balance_acc =balanced_accuracy_score(y_test, y_pred)
        accuracy = acc(y_test, y_pred)
        precision_score =prec(y_test, y_pred, average='macro')
        recall_score = rec(y_test, y_pred, average='macro')
        funo= f1(y_test, y_pred, average='macro')
        matriz = confusion_matrix(y_test, y_pred)
       
        
        print('Modelo:',re.findall(r'\b(\w+)\b\'\>',str(i))[0])
        print('balance accuracy:', balance_acc,'\naccuracy:',accuracy,'\nPrecision score',precision_score,'\nrecall score',
             recall_score,'\nf1',funo,'\nmatriz=\n',matriz, '\n')
    
```
