```python
string ='''@Ironhack's-#Q website 776-is http://ironhack.com [(2018)]")'''
```

```python
#https://stackoverflow.com/questions/11331982/how-to-remove-any-url-within-a-string-in-python
#https://stackoverflow.com/questions/817122/delete-digits-in-python-regex
#https://stackoverflow.com/questions/43358857/how-to-remove-special-characters-except-space-from-a-file-in-python/43358965
def clean_up(s):
    import re
    
    paso1 = s.lower()
    paso2 = re.sub(r'http\S+', '', paso1) #quita http
    paso3 = re.sub("\d+", "", paso2) #quita numeros
    
    for k in paso3.split("\n"):
    #print(re.sub(r"[^a-zA-Z0-9]+", ' ', k))
    # Or:
        final = " ".join(re.findall(r"[a-zA-Z0-9]+", k))
    

    return final
    
```

```python
uno = clean_up(string)
```

```python
def tokenize(s):
    return s.split()
```

```python
dos = tokenize(uno)
dos
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('was')

lemmatizer.lemmatize('runs', pos='v')
```

```python
def stem_and_lemmatize(l):
    from nltk.stem import WordNetLemmatizer
    from nltk.stem import PorterStemmer
    
    lista_steam = []
    lemmatizer = WordNetLemmatizer()
    ps = PorterStemmer()
    
    for i in l:
        y = lemmatizer.lemmatize(i)
        z= ps.stem(y)
        lista_steam.append(z)
        
    return lista_steam
```

```python
tres = stem_and_lemmatize(dos)
tres
```

```python
def remove_stopwords(l):
    from nltk.corpus import stopwords
    stopwords_en = stopwords.words('english') 
    
    
    clean_words = []
    for w in l:
        if (w not in stopwords_en) and len(w)>1:
            clean_words.append(w)
            
            
    return clean_words
```

```python
remove_stopwords(tres)
```

```python
data['text_processed']= data['text'].apply(clean_up)
data['text_processed'] =data['text_processed'].apply(tokenize)
```

```python
%%time

data['text_processed'] =data['text_processed'].apply(stem_and_lemmatize)
```

```python
%%time
data['text_processed'] =data['text_processed'].apply(remove_stopwords)
```

```python
data.head() #columna limpia
```

```python
#hago bag of words

bag_of_words= []
for i in data['text_processed']:
    bag_of_words += i
```

```python
bag_of_words[:100]
```

```python
len(bag_of_words)
```

```python
#importo
from nltk.probability import FreqDist
```

```python
#fecuencias
fdist = FreqDist(bag_of_words)
frecuencias = fdist.most_common(5000)
frecuencias
```

```python
#matriz
freq = [i[0] for i in frecuencias]
freq
```

```python
%%time
#lleno

matriz = data[['text_processed','target']]

matriz['is_positive'] = matriz['target'].apply(lambda x: True if x==4 else False)

matriz.drop('target', axis=1, inplace=True)

for i in freq:
    matriz[i] = matriz['text_processed'].apply(lambda x: True if i in x else False)
```

```python
matriz.head()
```

```python
%%time

lista = []
for i in range(len(matriz)):
    positivo = matriz.loc[i,'is_positive']
    dic = {}
    for e in freq:
        dic[e] = matriz.loc[i,e]
    lista.append((dic,positivo))
```

```python
print(len(lista))
```

```python
lista[0]
```

```python
training_set = lista[:24000]
testing_set = lista[24000:]
```

```python
#importo
from nltk.classify import NaiveBayesClassifier
from nltk.classify import accuracy 
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
nltk.classify.accuracy(classifier, testing_set)
```
